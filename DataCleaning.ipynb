```python
price = pd.DataFrame({'Day': range(1, 731)})
```

```python
price['Price_1'] = price['Day'].map(lambda x: purchase[purchase['Day'] == x]['Price_1'].values[0])
price['Price_2'] = price['Day'].map(lambda x: purchase[purchase['Day'] == x]['Price_2'].values[0])
price['Price_3'] = price['Day'].map(lambda x: purchase[purchase['Day'] == x]['Price_3'].values[0])
price['Price_4'] = price['Day'].map(lambda x: purchase[purchase['Day'] == x]['Price_4'].values[0])
price['Price_5'] = price['Day'].map(lambda x: purchase[purchase['Day'] == x]['Price_5'].values[0])
```

```python
def calRevenue(day):
    rev = 0
    filtered_data = purchase[purchase['Day'] == day]
    for index, row in filtered_data.iterrows():
        if row['Brand']!=0:
            rev += row['Quantity'] * row[f'Price_{int(row['Brand'])}']
    return rev
```

```python
price['Revenue'] = price['Day'].map(lambda x: calRevenue(x))
```

```python
price['Promotion_1'] = price['Day'].map(lambda x: purchase[purchase['Day'] == x]['Promotion_1'].values[0])
price['Promotion_2'] = price['Day'].map(lambda x: purchase[purchase['Day'] == x]['Promotion_2'].values[0])
price['Promotion_3'] = price['Day'].map(lambda x: purchase[purchase['Day'] == x]['Promotion_3'].values[0])
price['Promotion_4'] = price['Day'].map(lambda x: purchase[purchase['Day'] == x]['Promotion_4'].values[0])
price['Promotion_5'] = price['Day'].map(lambda x: purchase[purchase['Day'] == x]['Promotion_5'].values[0])
```

```python
def numPurchases(day, b):
    filtered_data = purchase[purchase['Day'] == day]
    rev=0
    for index, row in filtered_data.iterrows():
        if row['Brand']==b:
            rev += row['Quantity']
    return rev
```

```python
price['Sales_1'] = price['Day'].map(lambda x: int(numPurchases(x, 1)))
price['Sales_2'] = price['Day'].map(lambda x: int(numPurchases(x, 2)))
price['Sales_3'] = price['Day'].map(lambda x: int(numPurchases(x, 3)))
price['Sales_4'] = price['Day'].map(lambda x: int(numPurchases(x, 4)))
price['Sales_5'] = price['Day'].map(lambda x: int(numPurchases(x, 5)))
```

```python
price_elasticities = {}

for i in range(1, 6):
    # Calculate percentage change in quantity demanded
    minSales = price[f'Sales_{i}'].min()
    delta_sales = (price[f'Sales_{i}'].max() - price[f'Sales_{i}'].min()) 

    delta_price = (price[f'Price_{i}'].max() - price[f'Price_{i}'].min())

    # Calculate price elasticity
    elasticity = delta_sales / delta_price

    # Store the result
    price_elasticities[f'Brand_{i}'] = elasticity

# Print the price elasticities
for brand, elasticity in price_elasticities.items():
    print(f"Price elasticity for {brand}: {elasticity}")
```

```python
for i in range (1, 6):
    price[f'Csales_{i}'] = (price[f'Sales_{i}'] - price[f'Sales_{i}'].shift(1)) / price[f'Sales_{i}'].shift(1)
    price[f'Cprice_{i}'] = (price[f'Price_{i}'] - price[f'Price_{i}'].shift(1)) / price[f'Price_{i}'].shift(1)
    price[f'Csales_{i}']=price[f'Csales_{i}'].replace([np.inf, -np.inf], 0)
    price[f'Cprice_{i}']=price[f'Cprice_{i}'].replace([np.inf, -np.inf], 0)
price.fillna(0, inplace=True)
price
```

```python
brand_1=[]
brand_2=[]
cpe = []
for i in range (1, 6):
    for j in range (1,6):
            cross_price_elasticity = price[f'Csales_{i}'].mean()/price[f'Cprice_{j}'].mean()
            brand_1.append(i)
            brand_2.append(j)
            cpe.append(cross_price_elasticity)

df_cpe = pd.DataFrame({'Brand_1': brand_1, 'Brand_2': brand_2, 'CPE': cpe})
df_cpe
```
